```python
origin_date = pd.to_datetime('1899-12-30')
date_columns = ['Date submitted', 'Date received', 'Company_Response_Date']

for col in date_columns:
    df[col] = origin_date + pd.to_timedelta(df[col], unit='D')
```

```python
df.head()
```

```python
df = df.fillna({
    'Sub-product' : 'Not Specified' ,
    'Sub-issue' : 'Not Specified' ,
    'Company public response' : 'Not Specified' ,
    'Timely response?' : 'Not Specified'
})
```

```python
df_company = pd.read_excel("data/Consumer Financial Complaints Dataset.xlsx", sheet_name='Company')
```

```python
df_merged = pd.merge(df, df_company, on='Company_ID_1081', how='left')
```
